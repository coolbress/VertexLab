# YFinance Examples

Demonstrates fetching financial data using the YFinance provider.

**Prerequisites**: `pip install vertex-forager` (no API key required)

## 1. Basic: Fetch Price Data (In-Memory)

```python
from vertex_forager import create_client

client = create_client(provider="yfinance", rate_limit=60)
df = client.get_price_data(tickers=["AAPL", "MSFT"], show_progress=False)
df.head()
```

## 2. Persist to DuckDB

Pass `connect_db` with a `duckdb://` URI to write results to a local DuckDB file.
The return value changes from a DataFrame to a `RunResult` summary.

```python
res = client.get_price_data(
    tickers=["AAPL", "MSFT", "GOOGL"],
    connect_db="duckdb://./yfinance_demo.duckdb",
    show_progress=False,
)
print(res)
```

## 3. Multiple Datasets

YFinance supports several dataset types: `price`, `info`, `dividends`, `splits`, `actions`, `calendar`, `news`.

```python
# Fetch dividend history
dividends = client.get_data(
    tickers=["AAPL"],
    datasets=["dividends"],
    show_progress=False,
)
dividends.head()
```

## 4. Custom Configuration

Control concurrency, retry behavior, and flush thresholds via `create_client` kwargs.

```python
custom_client = create_client(
    provider="yfinance",
    rate_limit=30,
    concurrency=4,
    metrics_enabled=True,
)

res = custom_client.get_price_data(
    tickers=["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA"],
    connect_db="duckdb://./yfinance_custom.duckdb",
    show_progress=False,
)
print(res)
```
